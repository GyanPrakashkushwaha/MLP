### Q-11) According to the data which Locality have highest average Sale Price?

```python
locality = df.groupby(by='Locality')
```

```python
avg_sale_price = locality['Sale Price'].mean()
```

```python
max(avg_sale_price)
```

```python
highest_avg_sale_price = avg_sale_price.idxmax()
highest_avg_value = avg_sale_price.max()

print(f"Locality with the highest average sale price: {highest_avg_sale_price}")
print(f"Highest average sale price: {highest_avg_value}")
```

```python
locality['Sale Price'].mean() == 2270223.539495238
```

```python
locality['Sale Price'].mean()
```

```python
group_df = pd.DataFrame(locality['Sale Price'].mean())
```

```python
group_df['Locality'] = group_df.index
```

```python
group_df.reset_index(inplace=True,drop=True)
```

```python
group_df
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=group_df, x="Locality", y="Sale Price", palette="viridis")
plt.title("Average Sale Price by Locality", fontsize=14)
plt.xlabel("Locality", fontsize=12)
plt.ylabel("Average Sale Price", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import r2_score

# Given Data
actual_y = np.array([3, 7, 5, 9, 11])   # Actual values
y_pred = np.array([2.8, 6.5, 5.2, 8.7, 10.9])  # Predicted values

# Compute Mean of actual values
y_mean = np.mean(actual_y)

# Compute SS_tot (Total Variance)
SS_tot = np.sum((actual_y - y_mean) ** 2)

# Compute SS_res (Residual Variance)
SS_res = np.sum((actual_y - y_pred) ** 2)

# Compute R^2 Score manually
R2_manual = 1 - (SS_res / SS_tot)

# Compute R^2 Score using sklearn
R2_sklearn = r2_score(actual_y, y_pred)

# Print results
print(f"Mean of actual values: {y_mean}")
print(f"Total Variance (SS_tot): {SS_tot}")
print(f"Residual Variance (SS_res): {SS_res}")
print(f"R² Score (Manual Calculation): {R2_manual:.4f}")
print(f"R² Score (sklearn): {R2_sklearn:.4f}")

# Visualization
plt.figure(figsize=(8, 5))
sns.scatterplot(x=range(len(actual_y)), y=actual_y, label='Actual Values', marker='o', color='blue')
sns.scatterplot(x=range(len(y_pred)), y=y_pred, label='Predicted Values', marker='s', color='red')
plt.axhline(y=y_mean, color='gray', linestyle='--', label='Mean of Actual Values')

# Connect actual and predicted points to show error
for i in range(len(actual_y)):
    plt.plot([i, i], [actual_y[i], y_pred[i]], color='black', linestyle='dotted')

plt.xlabel("Index")
plt.ylabel("Values")
plt.title("Actual vs Predicted Values with Errors")
plt.legend()
plt.show()
```

```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.model_selection import validation_curve
from sklearn.linear_model import LogisticRegression
X, y = make_classification(n_samples=1_000, random_state=0)
logistic_regression = LogisticRegression()
param_name, param_range = "C", np.logspace(-8, 3, 10)
train_scores, test_scores = validation_curve(
    logistic_regression, X, y, param_name=param_name, param_range=param_range
)
print(f"The average train accuracy is {train_scores.mean():.2f}")
# The average train accuracy is 0.81
print(f"The average test accuracy is {test_scores.mean():.2f}")
# The average test accuracy is 0.81
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set_theme('paper')
import os
os.chdir(r'D:/Gyan Prakash Kushwaha/MLP/')
```

```python
df = pd.read_csv(r'Data/Social_Network_Ads.csv')
```

```python
df = df.iloc[:,2:]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Purchased',axis=1),
                                                    df['Purchased'],
                                                    test_size=0.3,
                                                    random_state=0)
```

```python
X_train.shape , X_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
normalize = MinMaxScaler()
normalize
```

```python
normalize.fit(X_train)
X_train_normalized = normalize.transform(X_train)
X_test_normalized = normalize.transform(X_test)
```

```python
normalize.data_max_ , normalize.data_min_ , normalize.data_range_
```

```python
X_train_normalized = pd.DataFrame(X_train_normalized,columns=X_train.columns)
X_test_normalized = pd.DataFrame(X_test_normalized,columns=X_test.columns)
```

```python
# Now visualization of the data
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Before Scaling')
sns.kdeplot(X_train['Age'], ax=ax1)
sns.kdeplot(X_train['EstimatedSalary'], ax=ax1)

# after scaling
ax2.set_title('After Standard Scaling')
sns.kdeplot(X_train_normalized['Age'], ax=ax2)
sns.kdeplot(X_train_normalized['EstimatedSalary'], ax=ax2)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Age Distribution Before Scaling')
sns.kdeplot(X_train['Age'], ax=ax1)

# after scaling
ax2.set_title('Age Distribution After Standard Scaling')
sns.kdeplot(X_train_normalized['Age'], ax=ax2)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Salary Distribution Before Scaling')
sns.kdeplot(X_train['EstimatedSalary'], ax=ax1)

# after scaling
ax2.set_title('Salary Distribution Standard Scaling')
sns.kdeplot(X_train_normalized['EstimatedSalary'], ax=ax2)
plt.show()
```
